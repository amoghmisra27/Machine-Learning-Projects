```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
entropy_list = []
for red in range(1, 50):
    black = 50 - red
    
    entro = - (red/50 * np.log2(red/50) + black/50 * np.log2(black/50))
    entropy_list.append(entro)
```

```python
plt.plot(entropy_list)
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(1000, centers=5, random_state=42)
```

```python
classes = np.unique(y)
```

```python
for cls in classes:
    y_dash = (y == cls).astype(int)
    plt.figure()
    plt.scatter(X[:, 0], X[:, 1], c=y_dash)
    plt.show()
```
